```python
import pandas as pd
import numpy as np
import nltk
import string
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv("../raw_data/input_data.csv")
df.drop(columns=["Unnamed: 0"],inplace=True)
df["fichier"]=df["fichier"].str[0:-17]
```

```python
df
```

```python
df.set_index("fichier",inplace = True)
df
```

```python
df["clean"]=df["translation"]
df
```

```python
df["clean"]=df["clean"].str.lower()
```

```python
string.punctuation
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(df["clean"])
X.toarray()
```

```python
list_word = vectorizer.get_feature_names()
len(list_word)
```

```python
df_vect = pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names())
df_vect
```

```python
ds_word = df_vect.sum().sort_values(ascending=False).head(100)
ds_word
```

```python
df_word = pd.DataFrame(ds_word, columns=["count"])
```

```python
df_word["len_word"] = [len(i) for i in df_word.index]
```

```python
df_word["len_word"].value_counts()
```

```python
df_word_l=df_word[df_word["len_word"]>2]
df_word_l
```

```python
df_word_l3=df_word[df_word["len_word"]==10]
df_word_l3
```
