```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import json
```

```python
monRepertoire = "Data"
fichiers = [f for f in listdir(monRepertoire) if isfile(join(monRepertoire, f))]
fichiers
```

```python
df_fichiers = pd.DataFrame(fichiers, columns = ['Names'])
df_fichiers
```

```python
data = []
for fichier in fichiers :
    a = monRepertoire + "/" + fichier
    with open(a) as mon_fichier:
        data.append(json.load(mon_fichier))

# df_translation = pd.DataFrame(data, columns = ['Translation'])
data = [data[i]["Translation"] for i in range(len(data))]
df_translation = pd.DataFrame({'fichier': fichiers, 'translation': data})
df_translation
```
